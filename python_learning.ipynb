```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ' , '_')
```

```python
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# we can categorized the age column 
labled = ['young','adult','middle','senior']
df['age_group'] = pd.qcut(df['age'] ,q = 4 , labels = labled)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14 ,
    'Weekly' : 7 ,
    'Monthly' : 30 ,
    'Quarterly' : 90 ,
    'Bi-Weekly' : 14 ,
    'Annually' : 365 ,
    'Every 3 Months' : 90 
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.head()
```

```python
df[['discount_applied' , 'promo_code_used']].head()
```

```python
# At first glance the above columns look like same so it both are same we can remove one to avoid redundancy
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used' , axis = 1)
```
